```python
import re
```

```python
test_text = "Hello, world. This, is a text."
test_text_result = re.split(r'(\s)', test_text)

print(test_text_result[:99])
```

```python
test_text_result = re.split(r'([,.]|\s)', test_text)

print(test_text_result)
```

```python
test_text_result = [item for item in test_text_result if item.strip()]

print(test_text_result)
   
```

```python
test_text = "Hello, world. Is this-- a text?"

test_text_result = re.split(r'([,.:;?_!"()\']|--|\s)', test_text)
test_text_result = [item.strip() for item in test_text_result if item.strip()]
print(test_text_result)
```
